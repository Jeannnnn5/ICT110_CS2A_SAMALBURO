Given the code below    

    text = 'Business Analytics'


## CHALLENGE


Write a program which repeatedly reads a **full name**, then does the following:
- strips the name of any trailing spaces
- capitalizes the first letter
- puts the rest into smaller letters


```python
while True:

    full_name = input("Enter full name (or 'x' to exit): ") # Read full name from the user


    if full_name.lower() == 'x':
        print("Exiting the program.")    # Check if the user wants to exit
        break


    processed_name = full_name.strip().capitalize()  # Strip trailing spaces, capitalize the first letter, and convert the rest to lowercase

    print("Processed name:", processed_name)  # Print the processed name


```

Write a program to read through a file and print the contents of the file (line by line) all in upper case.

from [PEW] E8-1

```python
try:

    with open('/content/sample_data/r', 'r') as file:

        for line in file:

            print(line.upper(), end='')
except FileNotFoundError:
    print("File not found.")
```

#### Exercise 2  

Write a program to read through a file and output the number of lines in a file.

adapted from [PEW]

```python
try:

    with open('/content/sample_data/r', 'r') as file:

        line_count = 0


        for line in file:

            line_count += 1

        print("Number of lines in the file:", line_count)
except FileNotFoundError:
    print("File not found.")
```
